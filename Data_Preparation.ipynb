```python
import pandas as pd
```

**ASSETS**

```python
df_assets = pd.read_excel("Datasets_Before_Cleaning.xlsx", header=2)
```

```python
df_assets
```

```python
# Rename the columns based on the 3rd and 4th rows in your Excel sheet

new_columns = [
    'End of Period',
    'Cash on Hand',
    'Due from Central Bank',
    'Due from Domestic Banks',
    'Cash Items in Process of Collection',
    'Foreign Currency on Hand and Balances Due from Banks',
    'Treasury Bills',
    'Treasury Bonds',
    'Govt. of Sri Lanka Securities (a)',
    'Other Investments (c)',
    'Local',
    'Imports',
    'Exports',
    'Overdrafts',
    'Loans and Advances',
    'Total Loans and Advances (b)',
    'Fixed Assets and Other Assets',
    'Total Assets or Liabilities',
    'Percentage of Liquid Assets to Demand Deposits (e)',
    'Percentage of Loans & Advances to Total Deposits',
]

df_assets.columns = new_columns

# Now your DataFrame 'df_assets' should have the correct headers
print(df_assets.head())
```

```python
# Remove the first two rows (index 0 and 1)
df_assets = df_assets.iloc[2:].copy()

# Reset the index of the DataFrame
df_assets.reset_index(drop=True, inplace=True) # Corrected line: Using df_assets.reset_index()
```

```python
# Create the tuples for the corrected MultiIndex columns
multi_columns = [
    ('End of Period', '',''),
    ('Cash on Hand', '',''),
    ('Due from Central Bank', '',''),
    ('Due from Domestic Banks', '',''),
    ('Cash Items in Process of Collection', '',''),
    ('Foreign Currency on Hand and Balances Due from Banks', '',''),
    ('Investments', 'Govt.of Sri Lanka','Treasury Bills'),
    ('Investments', 'Govt.of Sri Lanka','Treasury Bonds (b)'),
    ('Investments', 'Govt.of Sri Lanka', 'Other Govt. of Sri Lanka Securities'),
    ('Investments', '','Other Investments (c)'),
    ('Loans and Advances','Bills Purchased and Discounted', 'Local'),
    ('Loans and Advances','Bills Purchased and Discounted', 'Imports'),
    ('Loans and Advances','Bills Purchased and Discounted', 'Exports'),
    ('Loans and Advances','', 'Overdrafts'),
    ('Loans and Advances','', 'Loans'),
    ('Loans and Advances','', 'Total Loans and Advance (b)'),
    ('Fixed and Other Assets', '',''),
    ('Total Assets or Liabilities', '',''),
    ('Percentage of Liquid Assets to Demand Deposits','',''),
    ('Percentage of Loans & Advances to Total Deposits','','')
]

# Create the MultiIndex
df_assets.columns = pd.MultiIndex.from_tuples(multi_columns)

# Now your DataFrame 'df_assets' has the corrected MultiIndex
print(df_assets.columns)
```

```python
df_assets
```

```python
datetime_column = 'End of Period'
```

```python
print(df_assets[datetime_column].dtype)
```

```python
# First, try to convert using the most common formats
df_assets[datetime_column] = pd.to_datetime(df_assets[datetime_column], errors='coerce')
```

```python
print(df_assets[datetime_column].dtype)
```

```python
df_assets
```

```python
# Identify the columns that should be numeric (excluding the first, last, and second)
numeric_cols_to_convert = df_assets.columns[3:-1]

# Iterate through the identified columns and attempt to convert them to numeric
for col in numeric_cols_to_convert:
    try:
        # Use pd.to_numeric to convert to numeric, coercing errors to NaN
        df_assets[col] = pd.to_numeric(df_assets[col], errors='coerce')
    except Exception as e:
        print(f"Error converting column '{col}': {e}")

# Display the data types of the columns after the attempted conversion
print(df_assets.dtypes)
```

```python
# Identify the names of your last two percentage columns
percentage_col1 = df_assets.columns[-2]
percentage_col2 = df_assets.columns[-1]

# Convert the columns to strings before applying str.rstrip
df_assets[percentage_col1] = df_assets[percentage_col1].astype(str).str.rstrip('%').astype('float')
df_assets[percentage_col2] = df_assets[percentage_col2].astype(str).str.rstrip('%').astype('float')
```

```python
df_assets
```

```python
#Multicolumn can cause error during visualisation therefore combined the names and brought to one name
df_assets.columns = ['_'.join(filter(None, col)).strip() for col in df_assets.columns]
```

```python
print(df_assets.isnull().sum())
```

```python
#less number of rows have missing values therefore dropping those rows will not impact majorly for analysis but tht will be the ultimate decision 
df_assets_cleaned = df_assets.dropna()
print(df_assets_cleaned.isnull().sum())
```

```python
# Drop unwanted columns, which are sum and percentages
df_assets_cleaned = df_assets_cleaned.drop(columns=[
    'Loans and Advances_Total Loans and Advance (b)',
    'Total Assets or Liabilities',
    'Percentage of Liquid Assets to Demand Deposits',
    'Percentage of Loans & Advances to Total Deposits'
])
```

```python
# Specify the file path for your CSV file
csv_file_path = "assets_data_cleaned.csv"

# Export the DataFrame to a CSV file
df_assets_cleaned.to_csv(csv_file_path, encoding='utf-8', index=False)

print(f"DataFrame successfully exported to: {csv_file_path}")
```
